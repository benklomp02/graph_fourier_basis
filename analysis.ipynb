## 3. Greedy Algorithm for Undirected Graphs

In this section, we will compare the results of the greedy algorithm for undirected graphs with the exact algorithm on the same graph as above. For the sake of completeness, will plot different variations of the greedy algorithm, i.e., the greedy algorithm using different objective functions. For two disjoint vertex sets $A, B \in V$, let us denote 
$$
\mathbf{W}(A, B) = \sum_{i \in A, j \in B} w_{ij}
$$ as the weight of the edges between $A$ and $B$. The greedy algorithm will be run with the following objective functions:
1. $F(A, B) = \mathbf{W}(A, B) / (|A| \cdot |B|)$, the average weight of edges between $A$ and $B$.
2. $F(A, B) = \mathbf{W}(A, B) / (|A| + |B|)$, the average weight of edges between $A$ and $B$ normalized by the product of the sizes of $A$ and $B$.
3. $F(A, B) = \mathbf{W}(A, B) / \max(|A|, |B|)$, the average weight of edges between $A$ and $B$ normalized by the maximum size of $A$ and $B$.
4. $F(A, B) = \mathbf{W}(A, B) / \min(|A|, |B|)$, the average weight of edges between $A$ and $B$ normalized by the minimum size of $A$ and $B$.

## 5. Greedy but for different Definitions of Total Variation
In this section, we consider the minimisation problem of the total Laplacian variation, which is defined as follows:
$$
\mathbf{J}(\mathbf{x}) = \| \mathbf{L} \mathbf{x} \|_1,
$$
where $\mathbf{L = D - W}$ is the combinatorial Laplacian matrix of the graph. The problem does not have a closed form solution, neither does it have a simple greedy application. However, we propose the following surrogates of the total Laplacian variation:
1. Triangle Upper Bound
1. Majority Logic Split 
1. Median Split 
1. Mean Split
